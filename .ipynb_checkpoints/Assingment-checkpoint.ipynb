```python
import pandas as pd
import numpy as np
import warnings 
import re
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv("output.csv")
```

```python
df
```

```python
class data_wrangling:
    def __init__(self,df):
        self.df=df
    def roll_no(self):
        df["Roll No."]=df["Roll No."].str.lower()
        df['Roll No.'][df['Roll No.'].str.isdigit()==True]=[i[:4] for i in df['Roll No.'][df.Age==df.Age.min()].values if len(i)>3][0]+df['Roll No.'][df['Roll No.'].str.isdigit()==True].values[0]
        df['Roll No.']=df['Roll No.'].str.replace("e",'')
        df['Roll No.']=df['Roll No.'].str.replace("e-",'')
        df['Roll No.']=df['Roll No.'].str.replace("s-","s")
        df['Roll No.']=df['Roll No.'].str.replace(" ","")
        return df
    def Age(self):
        df['Age']=df['Age'].apply(lambda x: re.sub("[a-zA-z]","",x))
        return df
    def Gender(self):
        df['Gender']=df['Gender'].str.capitalize()
        df.Gender[df.Gender.str.isdigit() ==True] = df.Gender.mode()
        df['Gender'] = df['Gender'].fillna(df['Gender'].value_counts().idxmax())
        return df
    def city(self):
        df['City']=df.City.str.capitalize()
        df['City']=df.City.str.replace("Bwp","Bahawalpur")
        return df
    def income(self):
        df['Income']=df['Income'].apply(lambda x: re.sub("\W+","",x))
       #\W+ Matches any character that is not a word character(alphanumeric&underscores)
        return df
    def stipend(self):
        df['Stipend'] = df.Stipend.apply(lambda x: re.sub("(Ehsas)",str(44000),x))
        df['Stipend'] = df.Stipend.apply(lambda x: re.sub("[a-zA-Z].*",str(0),x))
        return df
    def study_hours(self):
        df['Study hours']=df['Study hours'].fillna('')
        df['Study hours'] = df['Study hours'].apply(lambda x: re.sub(".[a-zA-z]|[a-zA-z/+]","",x))
        return df
```

```python
#Making an object of class 
obj=data_wrangling(df)
#Calling Methods using object
obj.roll_no()
obj.Age()
obj.Gender()
obj.city()
obj.income()
obj.stipend()
obj.study_hours()
```
